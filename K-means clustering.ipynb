```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)  # 3 clusters for 3 iris species
kmeans.fit(X)

# Add cluster labels to the data
X['Cluster'] = kmeans.labels_
```

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X.drop('Cluster', axis=1))
X['PCA1'] = principal_components[:, 0]
X['PCA2'] = principal_components[:, 1]

plt.figure(figsize=(8,6))
sns.scatterplot(data=X, x='PCA1', y='PCA2', hue='Cluster', palette='Set1', s=100)
plt.title("K-Means Clustering on Iris Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title='Cluster')
plt.show()
```
